#Matplotlib()

There are many plotting libraries for Data Visualization. Matplotlib is one such library used in Machine Learning for analysis and visualisations.

##Figure

Create a figure instance which provides an empty canvas using figure()

`fig=plt.figure()`  

Use `add_axes(list)` method to add axes to the fig.  

There has to be 4 elements in the list corresponding to left, bottom, width and height of the figure.  

The list element must be between 0 and 1. So, we can have list=[0,0,1,1].  

This will not show the axes. Commonly used list is list=[0.1,0.1,0.8,0.8]  

ax=fig.add_axes([0,0,1,1])

Sometimes, the axes are not visible with this setting, then we can use the settings similar to the below:

ax=fig.add_axes([0.1,0.1,0.8,0.8]) #commonly used

##Line Plot

A line plot is a type of chart or graph which displays information as a series of data points connected by straight line segments.  

It can be plotted using matplotlib.pyplot.plot() function.  

`plt.plot(x,y, scalex=True, scaley=True, data=None, color=None)`

##Plot
Use `plot()` method of the axes object to create a plot.  

`ax.plot([x], y, [fmt])` #Plot y versus x.  

[fmt] decides the lines/markers type and colors.

`ax.plot(x, y, 'go-')` # green, circle marker, solid line

`ax.plot(x, y, 'rs-.')` # red, square marker, dash-dot line

Use legend() method of axes class to add legend to the plot.

`ax.legend(labels=('stringPlot1','stringPlot2'), loc=codeFromTable)`

```python
np.random.seed(31)
dataPoints=10
y1=np.random.normal(5,10,size=dataPoints)
y2=np.random.normal(500,20,size=dataPoints)
x=np.linspace(0,9,10)
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x, y1, 'bs--')
ax.plot(x, y2, 'r.-')
ax.set_title("Rectilinear Motion")
ax.set_xlabel('Time')
ax.set_ylabel('Height and Velocity')
ax.legend(labels=('Velocity','Height'), loc='best')
plt.show()
```

We can observe from the above two plots, the interpretation of the result is slightly difficult as we are plotting both parameters on the same y-axis. We can plot y-axis on both the sides of the plot by using twinx() method.

Following changes in #matplotlib Block will give the output as below

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x, y1, 'bs--')
ax2=ax.twinx()
ax2.plot(x, y2, 'r.-')
ax.set_title("Rectilinear Motion")
ax.set_xlabel('Time')
ax.set_ylabel('Velocity')
ax2.set_ylabel('Height')
ax.legend(labels=('Velocity',), loc=8)
ax2.legend(labels=('Height',), loc=5)
plt.show()
```

##Subplots
In case, we have more than two plots with varying range and values, the twinx() is not a valid solution. Matplotlib offers a method named subplot(), to have multiple plots on a figure.  

`ax1=fig.add_subplot('nrows','ncols','index')
ax1=plt.subplot('nrows','ncols','index')`

nrows, ncols and index are the number of rows, number of columns and the position of the subplot for a specified row and column respectively. We can give arguments as (1,2,1) or (121). Both have same result. For example, 121 means first subplot of row 1 and column 2. Similarly, 122 means second subplot of row 1 and column 2. When we create a plot without any subplot, it is implicit that the plot refers to a subplot(111)


```python
# Matplotlib Block
fig=plt.figure()
ax1=plt.subplot(121) #ax1=fig.add_subplot(121) will also work fine
ax1.set_xlabel("Time")
ax1.set_ylabel("Velocity")
ax1.plot(x, y1, 'b^-')
ax2=plt.subplot(122) #ax2=fig.add_subplot(122) will also work fine
ax2.set_xlabel("Time")
ax2.set_ylabel("Height")
ax2.plot(x, y2, 'r--')
fig.tight_layout(pad=2)#Run with Different pad value
plt.show()
```

```python
np.random.seed(31)
mu=3
n=50
x=np.random.normal(mu,1,size=n)
y=np.random.normal(mu,1,size=n)
z=np.random.normal(mu,1,size=n)
```

```python
plt.rcParams['figure.figsize']=(10,8)
```

```python
ax=plt.axes(projection="3d")
ax.scatter(x,y,z, s=100)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
```

```python
def function_z(x,y):
  return 50-(x**2+y**2)
```

```python
N=50
x_values=np.linspace(-5,5,N)
y_values=np.linspace(-5,5,N)
```

```python
X,Y=np.meshgrid(x_values,y_values)
```

```python
plt.scatter(X,Y)
```

```python
Z=function_z(X,Y)
```

```python
ax=plt.axes(projection="3d")
ax.plot_surface(X,Y,Z)
```
